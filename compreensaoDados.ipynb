### Correlação entre variáveis

```python
# Calcular a matriz de correlação
corr_matrix = dataset[['CREDIT_SCORE', 'ANNUAL_MILEAGE', 'SPEEDING_VIOLATIONS', 'DUIS', 'PAST_ACCIDENTS', 'OUTCOME']].corr()
print(corr_matrix)
```

```python
# Visualizar a matriz de correlação
plt.figure(figsize=(12, 8))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1, vmax=1,
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool)),
    square=True
)
plt.title("Matriz de Correlação entre Variáveis Numéricas")
plt.tight_layout()
plt.show()
```

### Relação entre variáveis categóricas

```python
# AGE vs OUTCOME
tabela_contagem = pd.crosstab(
    index=dataset['AGE'],
    columns=dataset['OUTCOME'],
    margins=True,
    margins_name="Total"
)

tabela_contagem.columns = ['Sem sinistro (0)', 'Com sinistro (1)', 'Total']
tabela_contagem.index.name = 'AGE'

tabela_contagem
```

```python
# AGE vs OUTCOME
tabela_contagem = pd.crosstab(
    index=dataset['AGE'],
    columns=dataset['OUTCOME'],
    margins=True,
    margins_name="Total",
    normalize='columns'  # Percentagem por coluna
).round(4)*100

tabela_contagem.columns = ['Sem sinistro (0)', 'Com sinistro (1)', 'Total']
tabela_contagem.index.name = 'AGE'

tabela_contagem
```

```python
# AGE vs OUTCOME
sns.countplot(x='AGE', hue='OUTCOME', data=dataset)
plt.title("Distribuição de OUTCOME por AGE")
plt.xlabel("AGE")
plt.ylabel("Número de Registos")
plt.legend(title='OUTCOME')
plt.show()
```
